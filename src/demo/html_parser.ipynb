```python
from bs4 import BeautifulSoup

html_text = ''

with open("test.html", 'r') as infile:
    html_text = infile.read()
    print(html_text)

def parse_announcement(html_text):
    soup = BeautifulSoup(html_text, 'html.parser')
    tables = soup.findAll('table') # 找到内容部分

    all_table_data = []
    for t in tables:
        rows = t.findAll('tr')
        table_data = []
        # parse <td>
        for r in rows:
            columns = r.findAll('td')
            col_text = [c.text for c in columns]
            table_data.append(col_text)
            print(', '.join(col_text))
        all_table_data.append(table_data)
    return {
        "tables": all_table_data
    }

def parse_bid_price(announcement):
    result = []
    for table in announcement['tables']:
        subject_index = None
        package_index = None
        price_index = None
        header = table[0]
        for i in range(0, len(header)):
            if is_subject(header[i]):
                subject_index = i
                continue
            if is_package(header[i]):
                package_index = i
                continue
            if is_max_price(header[i]):
                price_index = i
        if subject_index is None or package_index is None or price_index is None:
            subject_index = None
            package_index = None
            price_index = None
            continue
        print(f"{subject_index}, {package_index}, {price_index}")
        # 找到了中标信息表
        print(table)
        for i in range(1, len(table)):
            print(i)
            row = table[i]
            result.append({
                "subject": row[subject_index],
                "package": row[package_index],
                'price': row[price_index]
            })
    return result
        

def is_subject(text: str):
    return text.startswith('标的')

def is_package(text: str):
    return text.startswith('标包名称')

def is_max_price(text: str):
    return text.startswith('最高限价')

tables = parse_announcement(html_text)
result = parse_bid_price(tables)
print(result)
```
